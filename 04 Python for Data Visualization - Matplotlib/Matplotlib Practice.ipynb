
# MATPLOTLIB PRACTICE

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
x = np.arange(0,100)

y = x*2
z = x**2
```

```python
y[:5], z[:5]
```

```python
figure = plt.figure()
axes = figure.add_axes([0,0,1,1])

plt.plot(x,y)
plt.title("title")
plt.xlabel("x")
plt.ylabel("y")
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([.2,.5,.2,.2])

ax1.plot(x,y, color="red")
ax2.plot(y,x, color="blue")
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax1 = fig.add_axes([0.2,.5,.4,.4])
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.1,.5,.4,.4])

ax1.plot(x,z)
ax1.set_title("Larger")
ax1.set_xlabel('X')
ax1.set_ylabel('Z')
ax1.set_xlim(22)

ax2.plot(x,y)
ax2.set_title("Zoom")
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_xlim([20,22])
ax2.set_ylim([30,50])
```

```python
fig = plt.subplots(1,2)
```

```python
fig, axes = plt.subplots(1,2)
axes[0].plot(x,y, lw=2, ls='--')
axes[1].plot(x,z, color='red', lw=3)
```

```python
fig, axes = plt.subplots(2,1)
axes[0].plot(x,y, lw=2, ls='--')
axes[1].plot(x,z, color='red', lw=3)
```

```python
fig, axes = plt.subplots(1,2, figsize=(12,2))
axes[0].plot(x,y, lw=2, ls='--')
axes[1].plot(x,z, color='red', lw=3)
```
