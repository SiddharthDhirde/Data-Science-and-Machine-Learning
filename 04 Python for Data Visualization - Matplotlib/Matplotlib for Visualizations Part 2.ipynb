## Matplotlib 

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

axes[0].plot(x,y)
axes[0].set_title("First Plot")
axes[1].plot(y,x)
axes[1].set_title("SecondPlot")

plt.tight_layout() #avoid overlapping
```

## Figure Size

```python
fig = plt.figure(figsize=(3,2))

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig = plt.figure(figsize=(8,2))

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig, axes = plt.subplots(figsize=(8,2))

axes.plot(x,y)
```

```python
fig, axes = plt.subplots(ncols=1,nrows=2,figsize=(8,2))

axes[0].plot(x,y)
axes[1].plot(y,x)
```

```python
fig, axes = plt.subplots(ncols=1,nrows=2,figsize=(8,2))

axes[0].plot(x,y)
axes[1].plot(y,x)

fig.tight_layout()
```

```python
fig.savefig('Subplot Figure.png', dpi=400) # dpi - depth per inch
# fig.savefig('Subplot Figure.jpg')
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2)
ax.plot(x,x**3)
```
