## Seaborn Practice

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sns.set_style('whitegrid')
```

```python
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
sns.pairplot(titanic)
```

```python
titanic.head()
```

```python
sns.scatterplot(data=titanic, x='fare', y='age')
```

```python
sns.jointplot(x='fare', y='age', data=titanic)
```

```python
sns.histplot(titanic["fare"],bins=20,)
```

```python
sns.displot(data=titanic['fare'], color='red', bins=30)
```

```python
titanic.head()
```

```python
sns.boxplot(data=titanic,x='class',y="age")
```

```python
sns.boxplot(data=titanic,x='class',y="age", palette='rainbow')
```

```python
sns.swarmplot(data=titanic,x='class',y="age", hue='class')
```

```python
sns.swarmplot(data=titanic,x='class',y="age", palette='rainbow')
```

```python
sns.countplot(data=titanic, x="sex")
```

```python
sns.countplot(data=titanic, x="sex", palette='rainbow')
```

```python
sns.histplot(data=titanic, x='sex', hue='sex')
```

```python
titanic.corr(numeric_only=True)
```

```python
titanic_corr = titanic.corr(numeric_only=True)
sns.heatmap(data=titanic_corr, cmap='coolwarm')
```

```python
titanic_corr = titanic.corr(numeric_only=True)
sns.heatmap(data=titanic_corr, cmap='coolwarm')
plt.title("titanic.corr()")
```

```python
g = sns.FacetGrid(titanic, col='sex')
g.map(plt.hist, 'age')
```

```python
g = sns.FacetGrid(titanic, col='sex')
g.map(sns.distplot, 'age')
```
