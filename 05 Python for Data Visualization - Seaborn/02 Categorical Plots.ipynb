### Categorical Plots

```python
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
sns.barplot(data=tips, x='total_bill', y='sex')
```

```python
sns.barplot(data=tips, x='sex', y='total_bill')
```

```python
sns.barplot(data=tips, x='sex', y='total_bill',estimator='std')
```

```python
sns.countplot(x='sex', data=tips)
```

```python
sns.boxplot(x='day', y='total_bill', data=tips)
```

```python
sns.boxplot(x='day', y='total_bill', data=tips, hue='smoker')
```

```python
sns.violinplot(x='day', y='total_bill', data=tips)
```

```python
sns.boxplot(x='day', y='total_bill', data=tips)
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, hue='sex')
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, hue='sex' ,split=True)
```

```python
sns.stripplot(data=tips, x='day', y='total_bill')
```

```python
sns.stripplot(data=tips, x='day', y='total_bill',jitter=False)
```

```python
sns.stripplot(data=tips, x='day', y='total_bill', hue='sex')
```

```python
sns.stripplot(data=tips, x='day', y='total_bill', hue='sex')
```

```python
sns.swarmplot(data=tips, x='day', y='total_bill')
```

```python
sns.swarmplot(data=tips, x='day', y='total_bill', hue="sex")
```

```python
sns.swarmplot(data=tips, x='day', y='total_bill', hue="sex")
sns.violinplot(data=tips, x='day', y='total_bill', hue="sex")
```
