## Matrix Plots 

```python
import seaborn as sns
import numpy as np
%matplotlib inline
tips = sns.load_dataset('tips')
flights = sns.load_dataset('flights')
tips.head()
```

```python
flights.head()
```

```python
tips_correlation = tips.corr(numeric_only=True)
```

```python
sns.heatmap(tips_correlation)
```

```python
tips_correlation
```

```python
sns.heatmap(tips_correlation, annot=True)
```

```python
sns.heatmap(tips_correlation, annot=True, cmap='coolwarm')
```

```python
flights.head()
```

```python
flights.pivot_table(index='month', columns='year', values='passengers')
```

```python
fp = flights.pivot_table(index='month', columns='year', values='passengers')
sns.heatmap(fp)
```

```python
sns.heatmap(fp, cmap='magma', linecolor='white', linewidths=1)
```

```python
sns.heatmap(fp, cmap='magma', linecolor='white', linewidths=3)
```

```python
sns.heatmap(fp, cmap='coolwarm', linecolor='white', linewidths=1)
```

```python
sns.heatmap(fp, cmap='magma', linecolor='black', linewidths=1)
```
